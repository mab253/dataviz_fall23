```python
# what about creating a smaller DataFrame, just bill length vs species?

bill_length_species = penguins[["Bill Length (mm)", "Species"]]

bill_length_species.sort_values("Bill Length (mm)", ascending=False).head(10)
```

```python
# if we are going to visualize this, we could choose to get rid of NA outliers

bill_length_species = bill_length_species.dropna()
bill_length_species.shape

# more context on missing values: https://jakevdp.github.io/PythonDataScienceHandbook/03.04-missing-values.html
```

```python
# let's look @ the mean of the culmen length, by species

bill_length_species.groupby("Species").mean()

# documentation on groupby: https://pandas.pydata.org/docs/user_guide/groupby.html
```

```python
# let's start with a scatter plot of just this data, species + culmen length
# working with the matplotlib defaults ...

# we need to load matplotlib!
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(bill_length_species["Species"], bill_length_species["Bill Length (mm)"])
```

```python
# but if I want to show each species in a different color, I should filter my data

# make a list of all the unique values in the "Species" column, full names
species = bill_length_species["Species"].unique()
print(species)

# filter: just DataFrames of culmen length vs. species, one DataFrame per species
adelie = bill_length_species[bill_length_species["Species"] == species[0]]
gentoo = bill_length_species[bill_length_species["Species"] == species[1]]
chinstrap = bill_length_species[bill_length_species["Species"] == species[2]]

chinstrap.head()
```

```python
# remember: plt.scatter(x, y, options)
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

plt.scatter(adelie["Species"], adelie["Bill Length (mm)"], color="Orange", alpha = 0.5, s = 150)
plt.scatter(gentoo["Species"], gentoo["Bill Length (mm)"], color="Teal", alpha = 0.5, s = 150)
plt.scatter(chinstrap["Species"], chinstrap["Bill Length (mm)"], color="Purple", alpha = 0.5, s = 150)
```

```python
# to examine the mean along with the scatter, let's add a boxplot to each species
# matplotlib docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html#

y_data = [adelie["Bill Length (mm)"], gentoo["Bill Length (mm)"], chinstrap["Bill Length (mm)"]]

# remember the pattern! plt.type(data, options)
plt.boxplot(y_data, positions=[0, 1, 2])

# pattern: plt.scatter(x, y, options)
# to match the box plots to the scatter plots, we assign:
# adelie = 0, gentoo = 1, chinstrap = 2
# x values here need to be a list the same length as the y list
plt.scatter([0]*len(adelie), adelie["Bill Length (mm)"], color="Orange", alpha = 0.5, s = 100)
plt.scatter([1]*len(gentoo), gentoo["Bill Length (mm)"], color="Teal", alpha = 0.5, s = 100)
plt.scatter([2]*len(chinstrap), chinstrap["Bill Length (mm)"], color="Purple", alpha = 0.5, s = 100)
```

```python
# let's do some data transformation, and "jitter" the dots

plt.boxplot(y_data, positions=[0, 1, 2])

import random

# for each species, I want to have an x_data list, small random numbers in a range near their x value
# the list of x values needs to be the same length as the list of y values
x_adelie = [random.uniform(-0.1, 0.1) + 0for numbers in range(len(adelie))]
print(x_adelie)

x_gentoo = [random.uniform(-0.1, 0.1) + 1 for numbers in range(len(gentoo))]
x_chinstrap = [random.uniform(-0.1, 0.1) + 2 for numbers in range(len(chinstrap))]

plt.scatter(x_adelie, adelie["Bill Length (mm)"], color="Orange", alpha = 0.3, s = 80)
plt.scatter(x_gentoo, gentoo["Bill Length (mm)"], color="Teal", alpha = 0.3, s = 80)
plt.scatter(x_chinstrap, chinstrap["Bill Length (mm)"], color="Purple", alpha = 0.3, s = 80)
```

```python
# let's see a violin plot to this data
# documentation: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.violinplot.html

violins = plt.violinplot(
    y_data,
    positions=[0, 1, 2],
    widths=0.40,
    bw_method="silverman",
    showmeans=False,
    showmedians=False,
    showextrema=False
)
```

```python
# now we ADD the violin to the outside

# and style the median lines in the box plot
plt.boxplot(y_data, positions=[0, 1, 2], medianprops={"color": "Red", "linewidth": 2})

plt.scatter(x_adelie, adelie["Bill Length (mm)"], color="Orange", alpha = 0.3, s = 80)
plt.scatter(x_gentoo, gentoo["Bill Length (mm)"], color="Teal", alpha = 0.3, s = 80)
plt.scatter(x_chinstrap, chinstrap["Bill Length (mm)"], color="Purple", alpha = 0.3, s = 80)

violins = plt.violinplot(y_data,positions=[0, 1, 2],widths=0.40,bw_method="silverman")

# let's change the background fill to transparent - or, facecolor to 'none' for each violin plot
for body in violins['bodies']:
    body.set_facecolor('none')
    body.set_edgecolor('Black')
```

```python
# reproducing the plot ...
plt.figure(figsize=(14, 10))
# figsize = inches!

plt.boxplot(y_data, positions=[0, 1, 2], medianprops={"color": "Red", "linewidth": 2})

plt.scatter(x_adelie, adelie["Culmen Length (mm)"], color="Orange", alpha = 0.3, s = 80)
plt.scatter(x_gentoo, gentoo["Culmen Length (mm)"], color="Teal", alpha = 0.3, s = 80)
plt.scatter(x_chinstrap, chinstrap["Culmen Length (mm)"], color="Purple", alpha = 0.3, s = 80)

violins = plt.violinplot(y_data,positions=[0, 1, 2],widths=0.40,bw_method="silverman")

for body in violins['bodies']:
    body.set_facecolor('none')
    body.set_edgecolor('Black')

# but let's clean up the axes, titles, text, font, etc.!
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html
plt.rcParams["font.family"] = "monospace"
plt.xticks([0, 1, 2], ["Adelie", "Gentoo", "Chinstrap"], rotation=0, ha='center')
plt.ylabel("Culmen Length (mm)")
plt.xlabel("Penguin Species")
plt.title("\nDistribution of Bill Length Across Penguin Species\n", pad=10, fontsize=18)

# and save it
plt.savefig("violin.png", dpi=500, format="png")
```

```python
# nice!
# now ... let's go back to the larger penguin dataset
# and plot flipper length vs. body mass, for each of the 3 species

plt.scatter(penguins["Flipper Length (mm)"], penguins["Body Mass (g)"])
```

```python
# ok ... but we want to filter the dataset to get 3 species, differentiated

# first getting a smaller dataframe
flipper_mass = penguins[["Flipper Length (mm)", "Body Mass (g)", "Species"]]
flipper_mass.head()
```

```python
# removing na values ...
flipper_mass.dropna()
flipper_mass.shape
```

```python
# make filtered dataframes for each species: i.e., an adelie table with only flipper length and body mass

adelie_flip = flipper_mass[flipper_mass["Species"] == species[0]]
gentoo_flip = flipper_mass[flipper_mass["Species"] == species[1]]
chinstrap_flip = flipper_mass[flipper_mass["Species"] == species[2]]

adelie_flip.head(1)
```

```python
plt.figure(figsize=(10, 8))

# markers! https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers

plt.scatter(adelie_flip["Flipper Length (mm)"], adelie_flip["Body Mass (g)"], color = "Orange", marker="o", label = "adelie")
plt.scatter(gentoo_flip["Flipper Length (mm)"], gentoo_flip["Body Mass (g)"], color = "Teal", marker="v", label = "gentoo")
plt.scatter(chinstrap_flip["Flipper Length (mm)"], chinstrap_flip["Body Mass (g)"], color = "Purple", marker="s", label = "chinstrap")

# note: we could loop these, but we are spelling them out for now

plt.legend()

plt.ylabel("Body Mass (g)")
plt.xlabel("Flipper Length (mm)")
plt.title("\nFlipper Length and Body Mass\nfor 3 Penguin Species, Palmer\n", pad=6, fontsize=18)
```
