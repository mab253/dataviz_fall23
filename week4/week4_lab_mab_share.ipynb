```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
plate9 = pd.read_csv('https://raw.githubusercontent.com/ajstarks/dubois-data-portraits/master/plate09/data.csv')

plate9.head()
plate9 = plate9.rename(columns={'Negroes': 'Georgia'})
# this column name is a term that du Bois used in his own data,
# but a term I will re-name for the 2023 visualization
plate9.head()

# in my example I am going to use the 1800's data from du Bois, from AJ Starks, to make a python version of Plate 9:
# https://github.com/ajstarks/dubois-data-portraits/blob/master/plate09/original-plate-09.jpg
```

```python
from matplotlib import font_manager

font_manager.fontManager.addfont('/content/B52-ULC W00 ULC.ttf')

fig, (ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8) = plt.subplots(8, 1, sharex=True, figsize=(6, 7))
fig.set_facecolor("#dfd3c6")

fig.suptitle('AGE DISTRIBUTION OF BLACK PEOPLE IN GEORGIA\nCOMPARED WITH FRANCE', fontsize=22, y=1.05)

colors=["black", "#ffd700"]

axes = [ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8]
#for ax in [ax1, ax2, ax3, ax4, ax5]:

ax1.set_ylabel('\nAGES\nUNDER 10', labelpad=35, rotation='horizontal', fontsize="large", loc="center")
ax2.set_ylabel('\n10-20', labelpad=35, rotation='horizontal', fontsize="large")
ax3.set_ylabel('\n20-30', labelpad=35, rotation='horizontal', fontsize="large")
ax4.set_ylabel('\n30-40', labelpad=35, rotation='horizontal', fontsize="large")
ax5.set_ylabel('\n40-50', labelpad=35, rotation='horizontal', fontsize="large")
ax6.set_ylabel('\n50-60', labelpad=35, rotation='horizontal', fontsize="large")
ax7.set_ylabel('\n60-70', labelpad=35, rotation='horizontal', fontsize="large")
ax8.set_ylabel('\n70\nAND OVER', labelpad=35, rotation='horizontal', fontsize="large")

for i in range(len(axes)):

    ax = axes[i]

    ax.barh([0.4, 0], [plate9["Georgia"][i],plate9["France"][i]], height=0.4, align='center', color=colors)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_facecolor('#dfd3c6')

    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)

    x_placement = (plate9["Georgia"][i] / 2) - 1
    if (plate9["Georgia"][i] < 4):
      x_placement = 0

    x_placement_france = (plate9["France"][i] / 2) - 1
    if (plate9["France"][i] < 5):
      x_placement_france = 0

    ax.text(x_placement, 0.3, str(plate9["Georgia"][i]) + "%", color="white", fontsize=13)
    ax.text(x_placement_france, -0.1, str(plate9["France"][i]) + "%", color="black", fontsize=13)

    ax.yaxis.set_label_coords(-0.1, 0.2)


plt.rcParams['font.sans-serif'] = 'B52-ULC W00 ULC'
plt.rcParams['font.family'] = 'sans-serif'


```

```python
# i will show 1 more example, using AJ Starks' du Bois data, from Plate 51
# original: https://github.com/ajstarks/dubois-data-portraits/blob/master/plate51/original-plate-51.jpg
# showing proportion of free vs. enslaved people among african americans, 1790-1870

plate51 = pd.read_csv('https://raw.githubusercontent.com/ajstarks/dubois-data-portraits/master/plate51/data.csv')

plate51.head()
plate51.tail()
```

```python
x = plate51["Year"]
y = plate51["Free"]

fig, ax = plt.subplots()

plt.fill_between(x, y, color="#00aa00", alpha=0.9)
plt.fill_between(x, y, 100, color="black", alpha=0.9)
plt.plot(x, y, color="black", alpha=1)

ax = plt.gca()
ax.xaxis.tick_top()
ax.set_yticks([])
ax.invert_yaxis()

ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.text(0.4, 0.9, "LIBRE - FREE", transform=ax.transAxes, fontsize=15)
ax.text(0.5, 0.5, "SLAVES\nESCLAVES", transform=ax.transAxes, fontsize=15, color="white", ha='center')
ax.set_ymargin(0)

fig.set_facecolor('#dfd3c6')
ax.set_facecolor('#dfd3c6')
```
